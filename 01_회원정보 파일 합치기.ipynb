```python
# ê¸°ë³¸ ë¼ì´ë¸ŒëŸ¬ë¦¬
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()

# ê·¸ë˜í”„ ê¸°ë³¸ ì„¤ì •
plt.rcParams['font.family'] = 'Malgun Gothic'
# plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['figure.figsize'] = 12, 6
plt.rcParams['font.size'] = 14
plt.rcParams['axes.unicode_minus'] = False

# ê²½ê³  ëœ¨ì§€ ì•Šê²Œ
import warnings
warnings.filterwarnings('ignore')
```

```python
# 6ê°œ parquet íŒŒì¼ ê²½ë¡œ ë¦¬ìŠ¤íŠ¸ 
file_list = [
    'open/train/1.íšŒì›ì •ë³´/201807_train_íšŒì›ì •ë³´.parquet',
    'open/train/1.íšŒì›ì •ë³´/201808_train_íšŒì›ì •ë³´.parquet',
    'open/train/1.íšŒì›ì •ë³´/201809_train_íšŒì›ì •ë³´.parquet',
    'open/train/1.íšŒì›ì •ë³´/201810_train_íšŒì›ì •ë³´.parquet',
    'open/train/1.íšŒì›ì •ë³´/201811_train_íšŒì›ì •ë³´.parquet',
    'open/train/1.íšŒì›ì •ë³´/201812_train_íšŒì›ì •ë³´.parquet'
]

# ë¹ˆ ë°ì´í„°í”„ë ˆì„ ë¦¬ìŠ¤íŠ¸ ìƒì„±
dfs = []

# ê° íŒŒì¼ì„ ì½ì–´ì„œ dfs ë¦¬ìŠ¤íŠ¸ì— ì¶”ê°€
for file in file_list:
    df = pd.read_parquet(file)
    dfs.append(df)

# ë°ì´í„°í”„ë ˆì„ í•©ì¹˜ê¸° (í–‰ ê¸°ì¤€ìœ¼ë¡œ ì•„ë˜ë¡œ ë¶™ì´ê¸°)
merged_df = pd.concat(dfs, ignore_index=True)
```

```python
# ê²°ì¸¡ì¹˜ê°€ ìˆëŠ” ì»¬ëŸ¼ë§Œ ì¶œë ¥
for col in merged_df.columns:
    missing_count = merged_df[col].isnull().sum()
    if missing_count > 0:
        print(f"{col}: {missing_count}")
```

```python
# ì²´í¬í•  ê²°ì¸¡ì¹˜ ëŒ€ì²´ê°’ ë¦¬ìŠ¤íŠ¸
missing_values = [-99, -1]

# ê²°ê³¼ ì €ì¥ìš© ë¦¬ìŠ¤íŠ¸
results = []

for col in merged_df.columns:
    for val in missing_values:
        count = (merged_df[col] == val).sum()
        if count > 0:
            results.append({'column': col, 'missing_value': val, 'count': count})

# ê²°ê³¼ë¥¼ ë°ì´í„°í”„ë ˆì„ìœ¼ë¡œ ë³€í™˜
missing_summary = pd.DataFrame(results)

# ê²°ê³¼ ì¶œë ¥
print(missing_summary)
```
