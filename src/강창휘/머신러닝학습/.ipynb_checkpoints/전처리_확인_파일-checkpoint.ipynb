```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import gc

# 경고 메시지가 뜨지 않게 설정
import warnings 
warnings.filterwarnings('ignore')

# 그래프 설정
sns.set()

# 그래프 기본 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
# plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['figure.figsize'] = 12, 6
plt.rcParams['font.size'] = 14
plt.rcParams['axes.unicode_minus'] = False

# 인코딩
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
```

```python
df1 = pd.read_csv('data/회원정보_전처리.csv')
df2 = pd.read_csv('data/신용정보_전처리.csv')
df3 = pd.read_csv('data/성과정보_전처리.csv')
```

```python
na_box = []

for idx, val in df1.isna().sum().items() :
    # print(f'{idx}의 결측치 : {val}')
    if val > 0 :
        na_box.append(idx)
```

```python
na_box.pop(0)
```

```python
df1.drop(na_box, axis = 1, inplace = True)
```

```python
df2.select_dtypes(include = ['category', 'object'])
```

```python
df2['한도심사요청건수'] = df2['한도심사요청건수'].str.replace('회', '').str.replace('이상', '').astype('int')
```

```python
df2['한도증액횟수_R12M'] = df2['한도증액횟수_R12M'].str.replace('회', '').str.replace('이상', '').astype('int')
```

```python
obj_cols = ['카드론동의여부', 'RV전환가능여부']
df2_en = pd.get_dummies(df2[obj_cols]).astype('int')
```

```python
df2.drop(obj_cols, axis = 1, inplace = True)
```

```python
df2 = pd.concat([df2, df2_en], axis = 1)
```

```python
df1.to_csv('회원정보_preprocessing.csv', index = False)
df2.to_csv('신용정보_preprocessing.csv', index = False)
df3.to_csv('성과정보_preprocessing.csv', index = False)
```
