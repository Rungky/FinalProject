```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 깨짐 방지


# 절대 경로로 파일 불러오기
df = pd.read_csv(
    'C:/Users/tsmce/Desktop/workspace/final-project/test-2018_test_승인매출정보.csv',
    low_memory=False
)
```

### 전처리 및 파생 컬럼 생성

```python
# 1. 할부 관련 컬럼 분류
amount_cols = [col for col in df.columns if '할부금액' in col and 'R12M' in col]
count_cols = [col for col in df.columns if '할부건수' in col and 'R12M' in col]

# 2. 총합 컬럼 생성
df['총_할부금액'] = df[amount_cols].sum(axis=1)
df['총_할부건수'] = df[count_cols].sum(axis=1)

# 3. 유형별 금액 합계
df['총_유이자_금액'] = df[[col for col in amount_cols if '유이자' in col]].sum(axis=1)
df['총_무이자_금액'] = df[[col for col in amount_cols if '무이자' in col and '부분' not in col]].sum(axis=1)
df['총_부분무이자_금액'] = df[[col for col in amount_cols if '부분' in col]].sum(axis=1)

# 4. 할부 관련 컬럼 drop 함수
def preprocess_halbu(df):
    drop_missing = df.isnull().mean()[lambda x: x > 0.8].index.tolist()
    drop_zero = [col for col in df.columns if df[col].nunique() == 1 and df[col].unique()[0] == 0]
    drop_constant = [col for col in df.columns if df[col].nunique() == 1]
    drop_cols = list(set(drop_missing + drop_zero + drop_constant))
    drop_halbu_cols = [col for col in drop_cols if '할부' in col]
    df = df.drop(columns=drop_cols)
    return df, drop_cols, drop_halbu_cols
```

### 할부 개월 수별 평균 금액

```python
# 개월 수별 평균 금액
for period in ['3M', '6M', '12M', '14M']:
    period_cols = [col for col in amount_cols if f'_{period}_' in col]
    df[f'{period}_평균금액'] = df[period_cols].mean(axis=1)
```

### 그룹 분석 함수

```python
def analyze_group(df, group_name, col_pattern, target, agg='sum'):
    group_cols = [col for col in df.columns if col_pattern in col]
    
    # 결측치 & 단일값 제거
    drop_cols = df[group_cols].isnull().mean()[lambda x: x > 0.8].index.tolist()
    drop_cols += df[group_cols].nunique()[lambda x: x <= 1].index.tolist()
    group_cols = [col for col in group_cols if col not in drop_cols]
    
    # 그룹 컬럼 생성
    new_col = f'총_{group_name}' if agg == 'sum' else f'평균_{group_name}'
    df[new_col] = df[group_cols].agg(agg, axis=1)
    
    # 0 vs 0이 아닌 값 비교
    df[f'{group_name}_0여부'] = df[new_col] == 0
    sns.boxplot(x=f'{group_name}_0여부', y=target, data=df)
    plt.title(f"[{group_name}] 0 여부에 따른 {target} 분포")
    plt.xlabel(f"{new_col} == 0")
    plt.ylabel(target)
    plt.show()
    
    df.drop(columns=f'{group_name}_0여부', inplace=True)
```

### 함수 실행 예시

```python
# 유이자 금액이 0인 고객 vs 아닌 고객의 총 할부금액 비교
analyze_group(df, '유이자_금액', '할부금액_유이자', target='총_할부금액')

# 무이자 건수 분석
analyze_group(df, '무이자_건수', '할부건수_무이자', target='총_할부금액')

# 전체 건수 분석
analyze_group(df, '전체_건수', '할부건수_', target='총_할부금액')
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calculate_vif(X, thresh=10.0, verbose=True):
    """
    VIF가 thresh 이상인 변수를 반복적으로 제거합니다.
    X: 수치형 변수만 포함된 DataFrame
    thresh: 제거 기준 (기본값 10.0)
    verbose: 중간 로그 출력 여부
    """
    X = X.copy()
    dropped = []
    
    while True:
        vif = pd.Series(
            [variance_inflation_factor(X.values, i) for i in range(X.shape[1])],
            index=X.columns
        )
        
        max_vif = vif.max()
        if max_vif > thresh:
            drop_col = vif.idxmax()
            dropped.append((drop_col, max_vif))
            if verbose:
                print(f"❌ 제거: {drop_col} (VIF={max_vif:.2f})")
            X = X.drop(columns=[drop_col])
        else:
            break

    vif_df = pd.DataFrame({'변수': X.columns, 'VIF': [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]})
    return X, dropped, vif_df.sort_values(by='VIF', ascending=False)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from tqdm import tqdm

def vif_loop(X, threshold=100.0, verbose=True):
    """
    VIF가 threshold 이상인 변수만 반복적으로 제거하며 재계산합니다.
    진행률 표시 포함.
    
    Parameters:
        X (DataFrame): 수치형 변수만 포함된 데이터프레임
        threshold (float): 제거 기준 (기본값 100)
        verbose (bool): 제거 로그 출력 여부

    Returns:
        X_clean (DataFrame): 최종 남은 변수들
        removed (list): 제거된 변수와 VIF 기록
        final_vif (DataFrame): 최종 VIF 테이블
    """
    X = X.copy().dropna()
    removed = []

    while True:
        vif_vals = []
        for i in tqdm(range(X.shape[1]), desc="🔄 VIF 계산 중"):
            vif = variance_inflation_factor(X.values, i)
            vif_vals.append((X.columns[i], vif))

        vif_df = pd.DataFrame(vif_vals, columns=['변수', 'VIF']).sort_values(by='VIF', ascending=False)

        # 무한대 또는 threshold 이상인 변수 찾기
        high_vif = vif_df[vif_df['VIF'] >= threshold]
        if high_vif.empty:
            break

        # 가장 높은 VIF 변수 제거
        drop_var = high_vif.iloc[0]['변수']
        drop_vif = high_vif.iloc[0]['VIF']
        removed.append((drop_var, drop_vif))
        if verbose:
            print(f"❌ 제거: {drop_var} (VIF={drop_vif:.2f})")
        X = X.drop(columns=[drop_var])

    # 최종 VIF 재계산
    final_vif = pd.DataFrame({
        '변수': X.columns,
        'VIF': [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    }).sort_values(by='VIF', ascending=False)

    return X, removed, final_vif
```

```python
# 할부 관련 수치형 컬럼 추출
halbu_cols = [col for col in df.select_dtypes(include='number').columns if '할부' in col]
X = df[halbu_cols].dropna()

# VIF 무한대 또는 100 이상인 변수 제거 루프 실행
X_clean, removed_vars, final_vif = vif_loop(X, threshold=100.0)

# 제거된 변수 확인
print(f"\n🧹 제거된 변수 수: {len(removed_vars)}")
for var, vif in removed_vars:
    print(f" - {var} (VIF={vif:.2f})")

# 최종 VIF 테이블 출력
print("\n✅ 최종 남은 변수들의 VIF:")
print(final_vif)
```

```python
cluster_cols = [
    '이용금액_할부_B0M',
    '이용건수_할부_B0M',
    '이용금액_할부_R6M',
    '이용개월수_할부_R6M',
    '할부_경과일수'
]
```

```python
### 데이터 정규화
from sklearn.preprocessing import StandardScaler

df_cluster = df[cluster_cols].dropna().copy()
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_cluster)
```

```python
### 클러스터링 실행
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=42, n_init='auto')  # 군집 개수 조정 가능
df_cluster['cluster'] = kmeans.fit_predict(X_scaled)
```

```python
### 군집별 특성 요약 보기
# 군집별 평균값 요약
cluster_summary = df_cluster.groupby('cluster_label')[cluster_cols].mean().round(1)
display(cluster_summary)
```
