```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

# ‚ñ∂ Ïú†Ìú¥ Í¥ÄÎ†® Ïª¨Îüº Í∑∏Î£π Ï†ïÏùò
inactive_columns = ["ÎØ∏Ïù¥Ïö©_CA", "ÎØ∏Ïù¥Ïö©_Ìï†Î∂Ä", "ÎØ∏Ïù¥Ïö©_ÏùºÏãúÎ∂à"]
no_performance_columns = ["Ïó∞ÏÜçÎ¨¥Ïã§Ï†ÅÍ∞úÏõîÏàò_Í∏∞Î≥∏_24M_Ïπ¥Îìú", "Ïó∞ÏÜçÏú†Ïã§Ï†ÅÍ∞úÏõîÏàò_Í∏∞Î≥∏_24M_Ïπ¥Îìú"]
usage_count_columns = [
    "Ïù¥Ïö©Í±¥Ïàò_Ï≤¥ÌÅ¨_R3M", "Ïù¥Ïö©Í±¥Ïàò_Ï≤¥ÌÅ¨_R6M", "Ïù¥Ïö©Í±¥Ïàò_Ï≤¥ÌÅ¨_R12M",
    "Ïù¥Ïö©Í±¥Ïàò_Ïã†Ïö©_B0M", "Ïù¥Ïö©Í±¥Ïàò_Ïã†Ïö©_R12M",
    "Ïù¥Ïö©Í±¥Ïàò_Ïã†Ìåê_B0M", "Ïù¥Ïö©Í±¥Ïàò_Ïã†Ìåê_R3M", "Ïù¥Ïö©Í±¥Ïàò_Ïã†Ìåê_R6M", "Ïù¥Ïö©Í±¥Ïàò_Ïã†Ìåê_R12M",
    "Ïù¥Ïö©Í±¥Ïàò_ÏùºÏãúÎ∂à_B0M", "Ïù¥Ïö©Í±¥Ïàò_ÏùºÏãúÎ∂à_R3M", "Ïù¥Ïö©Í±¥Ïàò_ÏùºÏãúÎ∂à_R6M", "Ïù¥Ïö©Í±¥Ïàò_ÏùºÏãúÎ∂à_R12M",
    "Ïù¥Ïö©Í±¥Ïàò_CA_R12M", "Ïù¥Ïö©Í±¥Ïàò_Ìï†Î∂Ä_Î¨¥Ïù¥Ïûê_R12M",
    "Ïù¥Ïö©Í±¥Ïàò_Ïò§ÌîÑÎùºÏù∏_B0M", "Ïù¥Ïö©Í±¥Ïàò_Ïò§ÌîÑÎùºÏù∏_R3M", "Ïù¥Ïö©Í±¥Ïàò_Ïò§ÌîÑÎùºÏù∏_R6M",
    "Ïù¥Ïö©Í±¥Ïàò_ÏÑ†Í≤∞Ï†ú_R6M"
]
usage_gap_columns = [
    "Ïù¥Ïö©ÌõÑÍ≤ΩÍ≥ºÏõî_Ïã†Ïö©", "Ïù¥Ïö©ÌõÑÍ≤ΩÍ≥ºÏõî_Ïã†Ìåê", "Ïù¥Ïö©ÌõÑÍ≤ΩÍ≥ºÏõî_Ï≤¥ÌÅ¨",
    "ÏµúÏ¢ÖÏπ¥ÎìúÎ°†Ïù¥Ïö©Í≤ΩÍ≥ºÏõî", "RPÌõÑÍ≤ΩÍ≥ºÏõî_Ï†ÑÍ∏∞"
]

target_columns = inactive_columns + no_performance_columns + usage_count_columns + usage_gap_columns

# ‚ñ∂ Ïã§ÏÇ¨Ïö©/ÎØ∏ÏÇ¨Ïö©/Ï∑®ÏÜå ÌååÏÉù Ïª¨Îüº ÏÉùÏÑ±
def categorize_usage(x):
    if x < 0:
        return "Ï∑®ÏÜå"
    elif x == 0:
        return "ÎØ∏ÏÇ¨Ïö©"
    else:
        return "Ïã§ÏÇ¨Ïö©"

for col in target_columns:
    df[f"{col}_Î∂ÑÎ•ò"] = df[col].apply(categorize_usage)

# ‚ñ∂ Segment ÏÉâÏÉÅ Î∞è ÏàúÏÑú ÏÑ§Ï†ï
segment_palette = {
    "A": "#FF9999", "B": "#99CC99", "C": "#99CCCC", "D": "#FFCC99", "E": "#CC99CC"
}
segment_order = ["A", "B", "C", "D", "E"]

# ‚ñ∂ ÏÉòÌîå Ï∂îÏ∂ú
df_sampled, _ = train_test_split(df, stratify=df["Segment"], test_size=500, random_state=42)

# ‚ñ∂ Î≤îÏ£ºÌòï Ïù∏ÏΩîÎî©
encoded_sample = df_sampled[[f"{col}_Î∂ÑÎ•ò" for col in target_columns]].apply(LabelEncoder().fit_transform)

# ‚ñ∂ PCAÎ°ú 2Ï∞®Ïõê Ï∂ïÏÜå
pca_result = PCA(n_components=2, random_state=0).fit_transform(encoded_sample)
df_sampled["pca_x"] = pca_result[:, 0]
df_sampled["pca_y"] = pca_result[:, 1]

# ‚úÖ ‚ñ∂ KMeans ÌÅ¥Îü¨Ïä§ÌÑ∞ÎßÅ (Ïò§ÌÉÄ ÏàòÏ†ïÎê®!)
kmeans = KMeans(n_clusters=5, random_state=0)
df_sampled["cluster"] = kmeans.fit_predict(encoded_sample)  # ‚úÖ Ïó¨Í∏∞ Ïò§ÌÉÄ ÏàòÏ†ï: en„Ñ¥coded_sample ‚Üí encoded_sample

# ‚ñ∂ Í≤∞Í≥º ÏãúÍ∞ÅÌôî
plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=df_sampled,
    x="pca_x",
    y="pca_y",
    hue="Segment",
    hue_order=segment_order,
    palette=segment_palette,
    alpha=0.6
)
plt.title("Ïú†Ìú¥ ÏÉÅÌÉú Í∏∞Î∞ò Í≥†Í∞ù Î∂ÑÌè¨ ÏãúÍ∞ÅÌôî (PCA)", fontsize=14)
plt.legend(title="Segment", bbox_to_anchor=(1.02, 0.5), loc="center left")
plt.tight_layout()
plt.show()
```

### ÌÅ¥Îü¨Ïä§ÌÑ∞Î•º ÌÜµÌïú Ïù¥ÌÉà ÏòàÏ∏° Î™®Îç∏ ÏòàÏ∏°

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

# ‚ñ∂ Ïù¥ÌÉà Ïó¨Î∂Ä ÏÉùÏÑ± (Ïú†Ìú¥ ÌñâÎèô Í∏∞Ï§Ä)
df_sampled["Ïù¥ÌÉàÏó¨Î∂Ä"] = (
    (df_sampled["Ïó∞ÏÜçÎ¨¥Ïã§Ï†ÅÍ∞úÏõîÏàò_Í∏∞Î≥∏_24M_Ïπ¥Îìú"] >= 6) &
    (df_sampled["Ïù¥Ïö©ÌõÑÍ≤ΩÍ≥ºÏõî_Ïã†Ïö©"] >= 6)
).astype(int)

# ‚ñ∂ ÏòàÏ∏°ÏùÑ ÏúÑÌïú ÌäπÏÑ± Îç∞Ïù¥ÌÑ∞ÏÖã Íµ¨ÏÑ±
X = encoded_sample.copy()
X["cluster"] = df_sampled["cluster"]            # Íµ∞ÏßëÎ≤àÌò∏ Ìè¨Ìï®
X["Segment_ord"] = df_sampled["Segment"].map(    # SegmentÎ•º Ïà´ÏûêÌòïÏúºÎ°ú Î≥ÄÌôò (ÌïÑÏöî Ïãú)
    {"A": 0, "B": 1, "C": 2, "D": 3, "E": 4}
)

# ‚ñ∂ ÌÉÄÍπÉ Î≥ÄÏàò: Ïù¥ÌÉàÏó¨Î∂Ä
y = df_sampled["Ïù¥ÌÉàÏó¨Î∂Ä"]

# ‚ñ∂ ÌïôÏäµ/ÌÖåÏä§Ìä∏ Î∂ÑÌï†
X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y, random_state=42
)

# ‚ñ∂ Î™®Îç∏ ÏÉùÏÑ± Î∞è ÌïôÏäµ
model = RandomForestClassifier(n_estimators=100, random_state=0)
model.fit(X_train, y_train)

# ‚ñ∂ ÏòàÏ∏° Î∞è ÌèâÍ∞Ä
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
# Í∑∏Î£πÌôî Í∏∞Ï§Ä Î∞è ÏäπÏù∏ Ïã§Ï†Å Ìï≠Î™© Ï†ïÏùò
#  Î∂ÑÏÑù ÎåÄÏÉÅ ÏäπÏù∏ Ïã§Ï†Å Ïª¨Îüº
approve_metrics = [
    "Ïù¥Ïö©Í∏àÏï°_Ï≤¥ÌÅ¨_R12M",              # Ï≤¥ÌÅ¨Ïπ¥Îìú Ïã§Ï†Å
    "Ïù¥Ïö©Í∏àÏï°_ÏùºÏãúÎ∂à_R12M",           # Ïã†Ïö© ÏùºÏãúÎ∂à Ïã§Ï†Å
    "Ïù¥Ïö©Í∏àÏï°_Ìï†Î∂Ä_R12M",             # Ïã†Ïö© Ìï†Î∂Ä Ïã§Ï†Å
    "Ïù¥Ïö©Í±¥Ïàò_Ï≤¥ÌÅ¨_R12M",             # Ï≤¥ÌÅ¨Ïπ¥Îìú Í±¥Ïàò
    "Ïù¥Ïö©Í±¥Ïàò_Ïã†Ìåê_R12M"              # Ïã†Ïö© ÏäπÏù∏ Í±¥Ïàò
]


#  Í≥†Í∞ùÍµ∞ Î∂ÑÎ•ò Í∏∞Ï§Ä
group_keys = [
    "Segment",
    "Ïù¥Ïö©Í∏àÏï°ÎåÄ_ord",
    "Ïó∞ÏÜçÏú†Ïã§Ï†ÅÍ∞úÏõîÏàò_Í∏∞Î≥∏_24M_Ïπ¥Îìú",
    "RPÌõÑÍ≤ΩÍ≥ºÏõî_Ï†ÑÍ∏∞"
]

# Í∑∏Î£πÎ≥Ñ ÏäπÏù∏ Ïã§Ï†Å ÌèâÍ∑† ÏßëÍ≥Ñ
grouped_summary = (
    df_sampled
    .groupby(group_keys)[approve_metrics]
    .mean()
    .round(1)
    .reset_index()
)

#  Ï†ïÎ†¨ 
summary_sorted = grouped_summary.sort_values(by="Ïù¥Ïö©Í∏àÏï°_Ï≤¥ÌÅ¨_R12M", ascending=False)

# Í≤∞Í≥º ÎØ∏Î¶¨Î≥¥Í∏∞
summary_sorted.head()
```

```python
# Í≥†Í∞ù Îì±Í∏â ÏûêÎèô Î∂ÑÎ•ò or LTV ÏòàÏ∏° ÏãúÎèÑ
def Îì±Í∏âÎ∂ÑÎ•ò(row):
    if row["Ïã†Ïö©Ï¥ùÍ∏àÏï°_R12M"] >= 200000 and row["Ï†ÑÏ≤¥ÏäπÏù∏Í±¥Ïàò_R12M"] >= 300:
        return "Ïö∞Îüâ"
    elif row["Ïã†Ïö©Ï¥ùÍ∏àÏï°_R12M"] >= 80000:
        return "Ï§ëÍ∞Ñ"
    else:
        return "Í∏∞Ï¥à"

df_sampled["Í≥†Í∞ùÎì±Í∏â"] = df_sampled.apply(Îì±Í∏âÎ∂ÑÎ•ò, axis=1)
```

```python
# Í≥†Í∞ù Îì±Í∏âÎ≥Ñ ÌäπÏÑ± Î∂ÑÏÑù
# Í≥†Í∞ùÎì±Í∏â ÏàúÏÑú ÏßÄÏ†ï
df_sampled["Í≥†Í∞ùÎì±Í∏â_Ï†ïÎ†¨"] = pd.Categorical(
    df_sampled["Í≥†Í∞ùÎì±Í∏â"],
    categories=["Í∏∞Ï¥à", "Ï§ëÍ∞Ñ", "Ïö∞Îüâ"],
    ordered=True
)

#  Í∑∏Î£πÎ≥Ñ ÌèâÍ∑† ÌäπÏÑ± ÏßëÍ≥Ñ
grade_summary = (
    df_sampled
    .groupby("Í≥†Í∞ùÎì±Í∏â_Ï†ïÎ†¨")[["Ïã†Ïö©Ï¥ùÍ∏àÏï°_R12M", "Ï†ÑÏ≤¥ÏäπÏù∏Í±¥Ïàò_R12M", "Ïó∞ÏÜçÎ¨¥Ïã§Ï†ÅÍ∞úÏõîÏàò_Í∏∞Î≥∏_24M_Ïπ¥Îìú"]]
    .mean()
    .round(1)
)

#  Í≤∞Í≥º Î≥¥Í∏∞
grade_summary

```

```python
# Í≥†Í∞ù Îì±Í∏â ÏòàÏ∏° Î∂ÑÎ•ò Î™®Îç∏

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder

#  Segment Ïª¨Îüº ‚Üí Ïà´ÏûêÌòïÏúºÎ°ú Î≥ÄÌôò
le = LabelEncoder()
df_sampled["Segment_ord"] = le.fit_transform(df_sampled["Segment"])


#  ÏûÖÎ†• ÌäπÏÑ±
X = df_sampled[[
    "Ï†ÑÏ≤¥ÏäπÏù∏Í±¥Ïàò_R12M", "Ïã†Ïö©Ï¥ùÍ∏àÏï°_R12M",
    "Segment_ord", "Ïù¥Ïö©Í∏àÏï°ÎåÄ_ord",
    "Ïó∞ÏÜçÏú†Ïã§Ï†ÅÍ∞úÏõîÏàò_Í∏∞Î≥∏_24M_Ïπ¥Îìú", "RPÌõÑÍ≤ΩÍ≥ºÏõî_Ï†ÑÍ∏∞"
]]

#  ÌÉÄÍπÉ Î≥ÄÏàò
y = df_sampled["Í≥†Í∞ùÎì±Í∏â"]

#  ÌïôÏäµ/ÌÖåÏä§Ìä∏ Î∂ÑÌï†
X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y, random_state=42
)

#  Î™®Îç∏ ÌïôÏäµ
model = RandomForestClassifier(n_estimators=100, random_state=0)
model.fit(X_train, y_train)

# ‚ñ∂ ÏòàÏ∏° Î∞è ÌèâÍ∞Ä
y_pred = model.predict(X_test)
print(" Î∂ÑÎ•ò ÏÑ±Îä• Î¶¨Ìè¨Ìä∏:")
print(classification_report(y_test, y_pred))

# ‚ñ∂ confusion matrix ÏãúÍ∞ÅÌôî
cm = confusion_matrix(y_test, y_pred, labels=model.classes_)
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d', cmap="Blues", xticklabels=model.classes_, yticklabels=model.classes_)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title(" Í≥†Í∞ùÎì±Í∏â ÏòàÏ∏° ÌòºÎèô ÌñâÎ†¨")
plt.tight_layout()
plt.show()

# ‚ñ∂ Î≥ÄÏàò Ï§ëÏöîÎèÑ ÌôïÏù∏
importances = pd.Series(model.feature_importances_, index=X.columns).sort_values(ascending=False)
plt.figure(figsize=(8,5))
sns.barplot(x=importances.values, y=importances.index, palette="viridis")
plt.title(" Í≥†Í∞ùÎì±Í∏â ÏòàÏ∏° Î™®Îç∏: Î≥ÄÏàò ÏòÅÌñ•Î†•")
plt.xlabel("Importance")
plt.tight_layout()
plt.show()

```

```python
# XGBoost Î≥ÄÏàò ÏòÅÌñ•ÎèÑ : Í≤ΩÍ≥ºÏõîÏùò ÏòàÏ∏° Í∏∞Ïó¨ÎèÑ

from xgboost import XGBRegressor

X = df_sampled[[
    "Ï†ÑÏ≤¥ÏäπÏù∏Í±¥Ïàò_R12M", "Segment_ord", "Ïù¥Ïö©Í∏àÏï°ÎåÄ_ord",
    "Ïó∞ÏÜçÎ¨¥Ïã§Ï†ÅÍ∞úÏõîÏàò_Í∏∞Î≥∏_24M_Ïπ¥Îìú", "RPÌõÑÍ≤ΩÍ≥ºÏõî_Ï†ÑÍ∏∞", "Ï≤≠Íµ¨Í∏àÏï°_R3M"
]]
y = df_sampled["Ïã†Ïö©Ï¥ùÍ∏àÏï°_R12M"]

model = XGBRegressor(n_estimators=100, random_state=0)
model.fit(X, y)

importances = pd.Series(model.feature_importances_, index=X.columns).sort_values(ascending=False)
```

```python
# Î≥ÄÏàò ÏòÅÌñ•ÎèÑ ÏãúÍ∞ÅÌôî

plt.figure(figsize=(8,5))
sns.barplot(x=importances.values, y=importances.index, palette="viridis")
plt.title(" XGBoost Í∏∞Î∞ò LTV ÏòàÏ∏°: Î≥ÄÏàò ÏòÅÌñ•ÎèÑ")
plt.xlabel("Ï§ëÏöîÎèÑ")
plt.tight_layout()
plt.show()
```

```python
# Ï£ºÏöî Î≥ÄÏàò Í∏∞Î∞ò Ïö∞Îüâ Í≥†Í∞ù Ï∂îÏ∂ú
df_sampled["ÏòàÏ∏°_LTV"] = model.predict(X)
Ïö∞Îüâ_ÌõÑÎ≥¥Íµ∞ = df_sampled[df_sampled["ÏòàÏ∏°_LTV"] >= 400000]

print(" Ïö∞Îüâ Í≥†Í∞ù Ïàò:", len(Ïö∞Îüâ_ÌõÑÎ≥¥Íµ∞))
```

```python
# ÏòàÏ∏°Í∞í vs Ïã§Ï†úÍ∞í ÎπÑÍµê (ÏÇ∞Ï†êÎèÑ)

y_true = df_sampled["Ïã†Ïö©Ï¥ùÍ∏àÏï°_R12M"]
y_pred = df_sampled["ÏòàÏ∏°_LTV"]

plt.figure(figsize=(8,6))
sns.scatterplot(x=y_true, y=y_pred, alpha=0.4, color="teal")
plt.plot([y_true.min(), y_true.max()], [y_true.min(), y_true.max()], 'r--')
plt.xlabel("Ïã§Ï†ú Ïã†Ïö© ÏäπÏù∏ Í∏àÏï°")
plt.ylabel("ÏòàÏ∏° Ïã†Ïö© ÏäπÏù∏ Í∏àÏï°")
plt.title(" Ïã§Ï†ú vs ÏòàÏ∏° LTV")
plt.tight_layout()
plt.show()
```

```python
# ÌïµÏã¨ ÌäπÏÑ± Í∏∞Î∞ò Í≥†Í∞ù Îß§Ìä∏Î¶≠Ïä§ ÏÉùÏÑ±

df_sampled["Ï†ÑÎûµÍ∑∏Î£π"] = df_sampled["Segment_ord"].astype(str) + "_" + df_sampled["Ïù¥Ïö©Í∏àÏï°ÎåÄ_ord"].astype(str)

group_summary = df_sampled.groupby("Ï†ÑÎûµÍ∑∏Î£π")["ÏòàÏ∏°_LTV"].mean().sort_values(ascending=False)
print(group_summary.head(10))
```

```python
# Ï†ÑÎûµ Í∑∏Î£πÎ≥Ñ Í≥†Í∞ù Ïàò vs LTV Î∂ÑÌè¨ Ï†ïÎ¶¨

group_stats = df_sampled.groupby("Ï†ÑÎûµÍ∑∏Î£π").agg({
    "ÏòàÏ∏°_LTV": ["mean", "count"]
}).round(1).sort_values(("ÏòàÏ∏°_LTV", "mean"), ascending=False)

print(group_stats.head(10))
```

```python
# ÌïµÏã¨ Í∑∏Î£π Í≥†Í∞ù Ï∂îÏ∂ú -> Ïö∞Îüâ ÌÉÄÍπÉ Î¶¨Ïä§Ìä∏ ÏÉùÏÑ±

Ïö∞Îüâ_Í∑∏Î£πÎ™Ö = group_summary.head(1).index[0]  # LTV ÏµúÏÉÅÏúÑ Í∑∏Î£πÎ™Ö
Ïö∞ÎüâÍ≥†Í∞ù = df_sampled[df_sampled["Ï†ÑÎûµÍ∑∏Î£π"] == Ïö∞Îüâ_Í∑∏Î£πÎ™Ö]

print(f" Ïö∞Îüâ Í∑∏Î£πÎ™Ö: {Ïö∞Îüâ_Í∑∏Î£πÎ™Ö}, Í≥†Í∞ù Ïàò: {len(Ïö∞ÎüâÍ≥†Í∞ù)}")
```

```python
# Ï†ÑÎûµ Í∑∏Î£πÎ≥Ñ LTV ÏãúÍ∞ÅÌôî

plt.figure(figsize=(10,6))
sns.barplot(x=group_summary.index, y=group_summary.values, palette="rocket")
plt.title(" Ï†ÑÎûµ Í∑∏Î£πÎ≥Ñ ÌèâÍ∑† ÏòàÏ∏° LTV")
plt.xticks(rotation=45, ha='right')
plt.ylabel("ÌèâÍ∑† ÏòàÏ∏° LTV")
plt.tight_layout()
plt.show()
```

```python
# Î∂àÏùºÏπò Í∑∏Î£π ÌôïÏù∏ -> ÌòúÌÉù/Îì±Í∏â Ïû¨ÏÑ§Í≥Ñ Í∑ºÍ±∞ ÎèÑÏ∂ú

ÎπÑÍµê = df_sampled.groupby("Ï†ÑÎûµÍ∑∏Î£π")[["Ïã†Ïö©Ï¥ùÍ∏àÏï°_R12M", "ÏòàÏ∏°_LTV"]].mean().round(1)
ÎπÑÍµê["Ï∞®Ïù¥"] = ÎπÑÍµê["ÏòàÏ∏°_LTV"] - ÎπÑÍµê["Ïã†Ïö©Ï¥ùÍ∏àÏï°_R12M"]
ÎπÑÍµê.sort_values("Ï∞®Ïù¥", ascending=False).head(5)
```

### Ïπ¥Îìú Ïú†ÌòïÎ≥Ñ Ïã§Ï†Å vs ÏäπÏù∏Îß§Ï∂ú ÎπÑÍµê

```python
# Ïπ¥Îìú Ïú†ÌòïÎ≥Ñ Î≥ÄÏàòÏôÄ ÏäπÏù∏Îß§Ï∂ú ÍµêÍµê
#  Ï£ºÏöî Ïπ¥Îìú Ïú†ÌòïÎ≥Ñ Î≥ÄÏàò ÏÑ†ÌÉù
X_cardtype = df_sampled[[
    "Ïù¥Ïö©Í±¥Ïàò_Ï≤¥ÌÅ¨_R12M",
    "Ïù¥Ïö©Í±¥Ïàò_ÏùºÏãúÎ∂à_R12M",
    "Ïù¥Ïö©Í∏àÏï°_Ïπ¥ÎìúÎ°†_R12M",
    "Ïù¥Ïö©Í∏àÏï°_Ìï†Î∂Ä_Î¨¥Ïù¥Ïûê_R12M"
]]
y_sales = df_sampled["Ïã†Ïö©Ï¥ùÍ∏àÏï°_R12M"]


#  ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ ÌôïÏù∏
correlations = X_cardtype.corrwith(y_sales).round(3)
print(" Ïπ¥Îìú Ïú†ÌòïÎ≥Ñ Î≥ÄÏàòÏôÄ ÏäπÏù∏Îß§Ï∂ú Í∞Ñ ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ:\n", correlations)
```

```python
# ÌöåÍ∑ÄÎ∂ÑÏÑù : Ïπ¥Îìú Ïú†ÌòïÎ≥Ñ ÏòÅÌñ•Î†• ÏàòÏπòÌôî

from sklearn.linear_model import LinearRegression

X_cardtype = df_sampled[[
    "Ïù¥Ïö©Í±¥Ïàò_Ï≤¥ÌÅ¨_R12M",
    "Ïù¥Ïö©Í±¥Ïàò_ÏùºÏãúÎ∂à_R12M",
    "Ïù¥Ïö©Í∏àÏï°_Ïπ¥ÎìúÎ°†_R12M",
    "Ïù¥Ïö©Í∏àÏï°_Ìï†Î∂Ä_Î¨¥Ïù¥Ïûê_R12M"
]]
y_sales = df_sampled["Ïã†Ïö©Ï¥ùÍ∏àÏï°_R12M"]

model = LinearRegression()
model.fit(X_cardtype, y_sales)

coef_df = pd.DataFrame({
    "Î≥ÄÏàò": X_cardtype.columns,
    "ÌöåÍ∑ÄÍ≥ÑÏàò": model.coef_.round(1)
})
print(" Ïπ¥Îìú Ïú†ÌòïÎ≥Ñ ÏäπÏù∏Îß§Ï∂ú ÏòÅÌñ•Î†•:\n", coef_df)
```

```python
# ÏùºÏãúÎ∂à Ï§ëÏã¨ Í≥†Í∞ù vs Ï≤¥ÌÅ¨ Ï§ëÏã¨ Í≥†Í∞ù Î∂ÑÎ¶¨
df_sampled["ÏÜåÎπÑÏÑ±Ìñ•"] = df_sampled.apply(
    lambda row: "ÏùºÏãúÎ∂àÏ§ëÏã¨" if row["Ïù¥Ïö©Í±¥Ïàò_ÏùºÏãúÎ∂à_R12M"] > row["Ïù¥Ïö©Í±¥Ïàò_Ï≤¥ÌÅ¨_R12M"]
    else "Ï≤¥ÌÅ¨Ï§ëÏã¨", axis=1
)

#  Í∑∏Î£πÎ≥Ñ ÌèâÍ∑† ÏäπÏù∏Îß§Ï∂ú ÎπÑÍµê
df_sampled.groupby("ÏÜåÎπÑÏÑ±Ìñ•")["Ïã†Ïö©Ï¥ùÍ∏àÏï°_R12M"].mean().round(1)
```

```python
# ÏÜåÎπÑÏÑ±Ìñ• + ÏòàÏ∏° LTV + Í≤ΩÍ≥ºÏõî -> Í∞ÄÏπòÎäî ÎÜíÏßÄÎßå Ïù¥ÌÉà ÏúÑÌóòÏù¥ ÏûàÎäî Í≥†Í∞ù ÏãùÎ≥Ñ
Ïù¥ÌÉàÏúÑÌóò_Ïö∞ÎüâÍ≥†Í∞ù = df_sampled[
    (df_sampled["ÏÜåÎπÑÏÑ±Ìñ•"] == "ÏùºÏãúÎ∂àÏ§ëÏã¨") &
    (df_sampled["ÏòàÏ∏°_LTV"] >= 200000) &
    (df_sampled["RPÌõÑÍ≤ΩÍ≥ºÏõî_Ï†ÑÍ∏∞"] >= 6)
]

print(f" Ïù¥ÌÉà ÏúÑÌóò Ïö∞Îüâ Í≥†Í∞ù Ïàò: {len(Ïù¥ÌÉàÏúÑÌóò_Ïö∞ÎüâÍ≥†Í∞ù)}Î™Ö")
```

```python
# ÏãúÍ∞ÅÌôî : ÏòàÏ∏° LTV vs Í≤ΩÍ≥ºÏõî (ÏÜåÎπÑÏÑ±Ìñ•Î≥Ñ)

sns.scatterplot(
    data=df_sampled,
    x="RPÌõÑÍ≤ΩÍ≥ºÏõî_Ï†ÑÍ∏∞",
    y="ÏòàÏ∏°_LTV",
    hue="ÏÜåÎπÑÏÑ±Ìñ•",
    alpha=0.4
)

plt.title(" ÏÜåÎπÑÏÑ±Ìñ•Î≥Ñ ÏòàÏ∏° LTV vs Í≤ΩÍ≥ºÏõî")
plt.xlabel("Í≤ΩÍ≥ºÏõî")
plt.ylabel("ÏòàÏ∏° LTV")

#  Î≤îÎ°Ä ÏúÑÏπò Ï°∞Ï†ï
plt.legend(title="ÏÜåÎπÑÏÑ±Ìñ•", loc="center left", bbox_to_anchor=(1, 0.5))

plt.tight_layout()
plt.show()

```

```python
# Ïπ¥Îìú Ïú†ÌòïÎ≥Ñ ÏãúÍ∞ÅÌôî (ÏÇ∞Ï†êÎèÑ/Î∞ïÏä§ÌîåÎ°Ø)

for col in X_cardtype.columns:
    plt.figure(figsize=(6,4))
    sns.scatterplot(x=df_sampled[col], y=y_sales, alpha=0.4)
    plt.title(f" {col} vs ÏäπÏù∏Îß§Ï∂ú")
    plt.xlabel(col)
    plt.ylabel("Ïã†Ïö©Ï¥ùÍ∏àÏï°_R12M")
    plt.tight_layout()
    plt.show()
```

```python
# ÌååÏÉù Î≥ÄÏàò ÏÉùÏÑ±

# 3Í∞úÏõî ÏäπÏù∏Îß§Ï∂ú Ï¶ùÍ∞ÄÏú®

#  ÏßÅÏ†Ñ 3Í∞úÏõî Ï≤¥ÌÅ¨Ïπ¥Îìú ÏäπÏù∏Í∏àÏï°
Ïù¥Ï†Ñ3M_Ï≤¥ÌÅ¨ = df_sampled["Ïù¥Ïö©Í∏àÏï°_Ï≤¥ÌÅ¨_R6M"] - df_sampled["Ïù¥Ïö©Í∏àÏï°_Ï≤¥ÌÅ¨_R3M"]

#  Ï¶ùÍ∞ÄÏú® Í≥ÑÏÇ∞
df_sampled["ÏäπÏù∏Îß§Ï∂ú_Ï¶ùÍ∞ÄÏú®_3M"] = (
    df_sampled["Ïù¥Ïö©Í∏àÏï°_Ï≤¥ÌÅ¨_R3M"] - Ïù¥Ï†Ñ3M_Ï≤¥ÌÅ¨
) / Ïù¥Ï†Ñ3M_Ï≤¥ÌÅ¨
```

```python
# ÌååÏÉù Î≥ÄÏàò ÏÉùÏÑ±

# ÏÇ¨Ïö© Ï£ºÍ∏∞
#  ‚ë† Ïù¥ÏÉÅÏπò Ï†úÍ±∞: ÏäπÏù∏Í±¥Ïàò ‚â• 0 & Ïù¥Ïö©Í∞úÏõîÏàò > 0
df_cleaned = df_sampled[
    (df_sampled["Ï†ÑÏ≤¥ÏäπÏù∏Í±¥Ïàò_R12M"] >= 0) &
    (df_sampled["Ïù¥Ïö©Í∞úÏõîÏàò_Ïã†Ïö©_R12M"] > 0)
].copy()

#  ‚ë° ÏÇ¨Ïö©Ï£ºÍ∏∞ Í≥ÑÏÇ∞: ÏõîÌèâÍ∑† ÏäπÏù∏Í±¥Ïàò
df_cleaned["ÏÇ¨Ïö©Ï£ºÍ∏∞"] = df_cleaned["Ï†ÑÏ≤¥ÏäπÏù∏Í±¥Ïàò_R12M"] / df_cleaned["Ïù¥Ïö©Í∞úÏõîÏàò_Ïã†Ïö©_R12M"]

#  ‚ë¢ Í≤∞Í≥º ÌôïÏù∏
df_cleaned["ÏÇ¨Ïö©Ï£ºÍ∏∞"].describe()
```

```python
# Î≥ÄÏàòÎ≥Ñ Î∂ÑÌè¨ ÏãúÍ∞ÅÌôî -> ÏÜåÎπÑ ÌùêÎ¶Ñ ÌååÏïÖ
import seaborn as sns
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# ‚ñ∂ ÏÇ¨Ïö©Ï£ºÍ∏∞ Î∂ÑÌè¨
sns.histplot(df_cleaned["ÏÇ¨Ïö©Ï£ºÍ∏∞"], bins=30, kde=True, ax=axes[0], color="#0072B2")
axes[0].set_title(" ÏÇ¨Ïö©Ï£ºÍ∏∞ Î∂ÑÌè¨", fontsize=13)
axes[0].set_xlabel("ÏõîÌèâÍ∑† ÏäπÏù∏Í±¥Ïàò")
axes[0].set_ylabel("Í≥†Í∞ù Ïàò")

# ‚ñ∂ 3Í∞úÏõî ÏäπÏù∏Îß§Ï∂ú Ï¶ùÍ∞ÄÏú® Î∂ÑÌè¨
sns.histplot(df_cleaned["ÏäπÏù∏Îß§Ï∂ú_Ï¶ùÍ∞ÄÏú®_3M"], bins=30, kde=True, ax=axes[1], color="#D55E00")
axes[1].set_title(" 3Í∞úÏõî ÏäπÏù∏Îß§Ï∂ú Ï¶ùÍ∞ÄÏú® Î∂ÑÌè¨", fontsize=13)
axes[1].set_xlabel("Ï¶ùÍ∞ÄÏú® (ÏµúÍ∑º 3M ÎåÄÎπÑ ÏßÅÏ†Ñ 3M)")
axes[1].set_ylabel("Í≥†Í∞ù Ïàò")

plt.tight_layout()
plt.show()

```
