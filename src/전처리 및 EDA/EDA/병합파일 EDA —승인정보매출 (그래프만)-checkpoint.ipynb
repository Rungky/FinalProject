### 컬럼 설명

- 미사용 여부 -> 특정 카드 기능 / 업종을 아예 사용하지 않은 고객
- 사용 경과 기간 -> 마지막 사용 이후 시간이 많이 지났다면 휴면 가능성 존재
- 무실적 기간 -> 일정 기간 내 단 1건도 사용 실적이 없었던 고객
- 낮은 이용건수 -> 최근 기간동안 사용은 했지만 건수가 매우 적음


```python
### 파생 범주 생성 함수 정의
def categorize_usage(x):
    if x < 0:
        return "취소"
    elif x == 0:
        return "미사용"
    else:
        return "실사용"
```

```python
# 유휴 탐색용 컬럼 통합
target_columns = inactive_columns + no_performance_columns + usage_count_columns + usage_gap_columns

# 파생 범주형 컬럼 생성
for col in target_columns:
    df[f"{col}_분류"] = df[col].apply(categorize_usage)
```

```python
# 유휴 신호 개수 계산
df["유휴_징후_개수"] = df[[f"{col}_분류" for col in target_columns]].apply(
    lambda row: sum(row.isin(["미사용", "취소"])),
    axis=1
)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

# ▶ 유휴 관련 컬럼 그룹 정의
inactive_columns = ["미이용_CA", "미이용_할부", "미이용_일시불"]
no_performance_columns = ["연속무실적개월수_기본_24M_카드", "연속유실적개월수_기본_24M_카드"]
usage_count_columns = [
    "이용건수_체크_R3M", "이용건수_체크_R6M", "이용건수_체크_R12M",
    "이용건수_신용_B0M", "이용건수_신용_R12M",
    "이용건수_신판_B0M", "이용건수_신판_R3M", "이용건수_신판_R6M", "이용건수_신판_R12M",
    "이용건수_일시불_B0M", "이용건수_일시불_R3M", "이용건수_일시불_R6M", "이용건수_일시불_R12M",
    "이용건수_CA_R12M", "이용건수_할부_무이자_R12M",
    "이용건수_오프라인_B0M", "이용건수_오프라인_R3M", "이용건수_오프라인_R6M",
    "이용건수_선결제_R6M"
]
usage_gap_columns = [
    "이용후경과월_신용", "이용후경과월_신판", "이용후경과월_체크",
    "최종카드론이용경과월", "RP후경과월_전기"
]

target_columns = inactive_columns + no_performance_columns + usage_count_columns + usage_gap_columns

# ▶ 실사용/미사용/취소 파생 컬럼 생성
def categorize_usage(x):
    if x < 0:
        return "취소"
    elif x == 0:
        return "미사용"
    else:
        return "실사용"

for col in target_columns:
    df[f"{col}_분류"] = df[col].apply(categorize_usage)

# ▶ Segment 색상 및 순서 설정
segment_palette = {
    "A": "#FF9999", "B": "#99CC99", "C": "#99CCCC", "D": "#FFCC99", "E": "#CC99CC"
}
segment_order = ["A", "B", "C", "D", "E"]

# ▶ 샘플 추출
df_sampled, _ = train_test_split(df, stratify=df["Segment"], test_size=500, random_state=42)

# ▶ 범주형 인코딩
encoded_sample = df_sampled[[f"{col}_분류" for col in target_columns]].apply(LabelEncoder().fit_transform)

# ▶ PCA로 2차원 축소
pca_result = PCA(n_components=2, random_state=0).fit_transform(encoded_sample)
df_sampled["pca_x"] = pca_result[:, 0]
df_sampled["pca_y"] = pca_result[:, 1]

# ✅ ▶ KMeans 클러스터링 (오타 수정됨!)
kmeans = KMeans(n_clusters=5, random_state=0)
df_sampled["cluster"] = kmeans.fit_predict(encoded_sample)  # ✅ 여기 오타 수정: enㄴcoded_sample → encoded_sample

# ▶ 결과 시각화
plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=df_sampled,
    x="pca_x",
    y="pca_y",
    hue="Segment",
    hue_order=segment_order,
    palette=segment_palette,
    alpha=0.6
)
plt.title("유휴 상태 기반 고객 분포 시각화 (PCA)", fontsize=14)
plt.legend(title="Segment", bbox_to_anchor=(1.02, 0.5), loc="center left")
plt.tight_layout()
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

# ▶ 이탈 여부 생성 (유휴 행동 기준)
df_sampled["이탈여부"] = (
    (df_sampled["연속무실적개월수_기본_24M_카드"] >= 6) &
    (df_sampled["이용후경과월_신용"] >= 6)
).astype(int)

# ▶ 예측을 위한 특성 데이터셋 구성
X = encoded_sample.copy()
X["cluster"] = df_sampled["cluster"]            # 군집번호 포함
X["Segment_ord"] = df_sampled["Segment"].map(    # Segment를 숫자형으로 변환 (필요 시)
    {"A": 0, "B": 1, "C": 2, "D": 3, "E": 4}
)

# ▶ 타깃 변수: 이탈여부
y = df_sampled["이탈여부"]

# ▶ 학습/테스트 분할
X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y, random_state=42
)

# ▶ 모델 생성 및 학습
model = RandomForestClassifier(n_estimators=100, random_state=0)
model.fit(X_train, y_train)

# ▶ 예측 및 평가
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

| 연속무실적개월수_24M_카드  | ≈0             | 유휴 징후 → 장기 소비 예측에는 미미한 영향     |


```python
## 예측값 VS 실제값 시각화 (scatter plot)

plt.figure(figsize=(8,6))
sns.scatterplot(x=y_test, y=y_pred, alpha=0.4, color="teal")
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')  # 기준선
plt.xlabel("실제 신용 승인 금액")
plt.ylabel("예측 신용 승인 금액")
plt.title(" 실제값 vs 예측값 (LTV 예측 모델)")
plt.tight_layout()
plt.show()
```

### 경과월(최근이용일)과 승인매출의 관계 / 고객생애가치 LTV(Life time Value) 

```python
# 경과월별 승인매출 평균 비교
#  경과월 구간화
df_sampled["경과월_구간"] = pd.cut(
    df_sampled["RP후경과월_전기"],
    bins=[-1, 3, 6],
    labels=["0~3", "4~6"]
)

#  구간별 평균 승인매출 비교
df_sampled.groupby("경과월_구간")[["신용총금액_R12M", "전체승인건수_R12M"]].mean().round(1)
```

```python
# 선형회귀분석으로 경과월 영향도 확인
from sklearn.linear_model import LinearRegression

#  입력 변수: 경과월
X = df_sampled[["RP후경과월_전기"]]
y = df_sampled["신용총금액_R12M"]

#  모델 학습
model = LinearRegression()
model.fit(X, y)

#  회귀계수 확인
coef = model.coef_[0]
intercept = model.intercept_
print(f" 회귀계수: {round(coef, 2)}")
print(f" 절편: {round(intercept, 1)}")
```

```python
# 시각화 : 경과월 vs 승인매출

plt.figure(figsize=(8,5))
sns.boxplot(data=df_sampled, x="경과월_구간", y="신용총금액_R12M", palette="Set2")
plt.title(" 경과월 구간별 승인매출 분포")
plt.xlabel("최근 사용 경과월 구간")
plt.ylabel("신용 승인 금액 (12M)")
plt.ticklabel_format(style='plain', axis='y')
plt.tight_layout()
plt.show()

```

```python
# 회귀분석 : 경과월이 승인매출에 미치는 영향

from sklearn.linear_model import LinearRegression

X = df_sampled[["RP후경과월_전기"]]
y = df_sampled["신용총금액_R12M"]

model = LinearRegression()
model.fit(X, y)

print(" 회귀계수:", round(model.coef_[0], 2))
print(" 절편:", round(model.intercept_, 1))
```

```python
# XGBoost 변수 영향도 : 경과월의 예측 기여도

from xgboost import XGBRegressor

X = df_sampled[[
    "전체승인건수_R12M", "Segment_ord", "이용금액대_ord",
    "연속무실적개월수_기본_24M_카드", "RP후경과월_전기", "청구금액_R3M"
]]
y = df_sampled["신용총금액_R12M"]

model = XGBRegressor(n_estimators=100, random_state=0)
model.fit(X, y)

importances = pd.Series(model.feature_importances_, index=X.columns).sort_values(ascending=False)
```

```python
# 변수 영향도 시각화

plt.figure(figsize=(8,5))
sns.barplot(x=importances.values, y=importances.index, palette="viridis")
plt.title(" XGBoost 기반 LTV 예측: 변수 영향도")
plt.xlabel("중요도")
plt.tight_layout()
plt.show()
```

```python
# 주요 변수 기반 우량 고객 추출
df_sampled["예측_LTV"] = model.predict(X)
우량_후보군 = df_sampled[df_sampled["예측_LTV"] >= 400000]

print(" 우량 고객 수:", len(우량_후보군))
```

```python
# 예측값 vs 실제값 비교 (산점도)

y_true = df_sampled["신용총금액_R12M"]
y_pred = df_sampled["예측_LTV"]

plt.figure(figsize=(8,6))
sns.scatterplot(x=y_true, y=y_pred, alpha=0.4, color="teal")
plt.plot([y_true.min(), y_true.max()], [y_true.min(), y_true.max()], 'r--')
plt.xlabel("실제 신용 승인 금액")
plt.ylabel("예측 신용 승인 금액")
plt.title(" 실제 vs 예측 LTV")
plt.tight_layout()
plt.show()
```

| 구간 위치          | 의미                                 |
|--------------------|--------------------------------------|
| 🔴 빨간 대각선 근처 | 예측값 ≈ 실제값 → 정확도 높은 예측    |
| 🔻 기준선 아래쪽    | 과소예측 → 실제보다 예측값 낮음       |
| 🔺 기준선 위쪽      | 과대예측 → 실제보다 예측값 높음       |
| 🌐 전체 분포 형태   | 대부분 점들이 기준선 부근에 밀집됨    |


✔️ 빨간 기준선을 중심으로 점들이 고르게 분포 → 예측 모델이 실제 승인금액을 매우 안정적으로 복원

✔️ 극단적으로 벗어나는 점이 없다는 것도 모델 과적합이 아니고 현실적인 범위에서 잘 작동한다는 뜻

### 카드 유형별 실적 vs 승인매출 비교

```python
# 카드 유형별 변수와 승인매출 교교
#  주요 카드 유형별 변수 선택
X_cardtype = df_sampled[[
    "이용건수_체크_R12M",
    "이용건수_일시불_R12M",
    "이용금액_카드론_R12M",
    "이용금액_할부_무이자_R12M"
]]
y_sales = df_sampled["신용총금액_R12M"]


#  상관관계 확인
correlations = X_cardtype.corrwith(y_sales).round(3)
print(" 카드 유형별 변수와 승인매출 간 상관관계:\n", correlations)
```

```python
# 회귀분석 : 카드 유형별 영향력 수치화

from sklearn.linear_model import LinearRegression

X_cardtype = df_sampled[[
    "이용건수_체크_R12M",
    "이용건수_일시불_R12M",
    "이용금액_카드론_R12M",
    "이용금액_할부_무이자_R12M"
]]
y_sales = df_sampled["신용총금액_R12M"]

model = LinearRegression()
model.fit(X_cardtype, y_sales)

coef_df = pd.DataFrame({
    "변수": X_cardtype.columns,
    "회귀계수": model.coef_.round(1)
})
print(" 카드 유형별 승인매출 영향력:\n", coef_df)
```

```python
# 일시불 중심 고객 vs 체크 중심 고객 분리
df_sampled["소비성향"] = df_sampled.apply(
    lambda row: "일시불중심" if row["이용건수_일시불_R12M"] > row["이용건수_체크_R12M"]
    else "체크중심", axis=1
)

#  그룹별 평균 승인매출 비교
df_sampled.groupby("소비성향")["신용총금액_R12M"].mean().round(1)
```

```python
# 소비성향 + 예측 LTV + 경과월 -> 가치는 높지만 이탈 위험이 있는 고객 식별
이탈위험_우량고객 = df_sampled[
    (df_sampled["소비성향"] == "일시불중심") &
    (df_sampled["예측_LTV"] >= 200000) &
    (df_sampled["RP후경과월_전기"] >= 6)
]

print(f" 이탈 위험 우량 고객 수: {len(이탈위험_우량고객)}명")
```

```python
# 파생 변수 생성

# 3개월 승인매출 증가율

#  직전 3개월 체크카드 승인금액
이전3M_체크 = df_sampled["이용금액_체크_R6M"] - df_sampled["이용금액_체크_R3M"]

#  증가율 계산
df_sampled["승인매출_증가율_3M"] = (
    df_sampled["이용금액_체크_R3M"] - 이전3M_체크
) / 이전3M_체크
```

```python
# 파생 변수 생성

# 사용 주기
#  ① 이상치 제거: 승인건수 ≥ 0 & 이용개월수 > 0
df_cleaned = df_sampled[
    (df_sampled["전체승인건수_R12M"] >= 0) &
    (df_sampled["이용개월수_신용_R12M"] > 0)
].copy()

#  ② 사용주기 계산: 월평균 승인건수
df_cleaned["사용주기"] = df_cleaned["전체승인건수_R12M"] / df_cleaned["이용개월수_신용_R12M"]

#  ③ 결과 확인
df_cleaned["사용주기"].describe()
```
