```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 깨짐 방지


# 절대 경로로 파일 불러오기
df = pd.read_csv(
    'C:/Users/tsmce/Desktop/workspace/final-project/test-2018_test_승인매출정보.csv',
    low_memory=False
)
```

### 결측치 제거 및 데이터 정제

```python
# 1. 컬럼명에 'CA'가 포함된 컬럼만 선택
ca_columns = [col for col in df.columns if 'CA' in col]

# 2. 해당 컬럼들의 결측치 수와 비율 계산
missing_counts = df[ca_columns].isnull().sum()
missing_ratios = (missing_counts / len(df)) * 100

# 3. 결과를 하나의 데이터프레임으로 정리
missing_summary = pd.DataFrame({
    '결측치_갯수': missing_counts,
    '결측치_비율(%)': missing_ratios.round(2)
})

print(missing_summary)
```

```python
# 1. 각 컬럼별로 값이 0인 개수 계산
zero_value_counts = (df[ca_columns] == 0).sum()

# 2. 전체 행 수로 비율 계산
zero_value_ratios = (zero_value_counts / len(df)) * 100

# 3. 결과 정리
zero_summary = pd.DataFrame({
    '값이_0인_갯수': zero_value_counts,
    '값이_0인_비율(%)': zero_value_ratios.round(2)
})

print(zero_summary)
```

```python
# 결과 담을 딕셔너리 초기화
usage_summary = {}

# 각 컬럼마다 0 / 0 아닌 값 수와 비율 계산
for col in ca_columns:
    total = len(df)
    zero_count = (df[col] == 0).sum()
    nonzero_count = total - zero_count
    usage_summary[col] = {
        '0_값_갯수': zero_count,
        '0_비율(%)': round(zero_count / total * 100, 2),
        '사용_이력_갯수': nonzero_count,
        '사용_이력_비율(%)': round(nonzero_count / total * 100, 2)
    }

# 데이터프레임으로 변환
usage_df = pd.DataFrame.from_dict(usage_summary, orient='index')
print(usage_df)
```

```python
# 사용 이력 비율 시각화 (비율 높은 순 정렬)
usage_df.sort_values('사용_이력_비율(%)')['사용_이력_비율(%)'].plot(kind='barh', figsize=(8,6), title='CA 컬럼별 사용 이력 비율')
plt.xlabel('비율 (%)')
plt.tight_layout()
plt.show()
```

### 상관계수 및 VIF 값 파악

```python
# CA 관련 컬럼 필터링
ca_df = df[ca_columns].copy()

# 상관계수 행렬 계산
corr_matrix = ca_df.corr().abs()  # 절댓값 기준으로 상관관계 강도 보기

# 상삼각 행렬만 추출 (자기 자신 제외)
upper_tri = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# 쌍별로 정리 후 정렬
corr_pairs = (
    upper_tri.stack()
    .reset_index()
    .rename(columns={'level_0': '변수1', 'level_1': '변수2', 0: '상관계수'})
    .sort_values(by='상관계수', ascending=False)
)

# 결과 출력
print("✅ CA 관련 변수 간 상관계수 (상위 항목):")
print(corr_pairs)
```

```python
X = df[ca_columns]

from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

# 상수항 추가
X_const = add_constant(X)

# VIF 계산
vif_df = pd.DataFrame()
vif_df["컬럼"] = X.columns
vif_df["VIF"] = [variance_inflation_factor(X_const.values, i+1) for i in range(len(X.columns))]  # i+1: 상수항 제외

# VIF 높은 순 정렬
vif_df = vif_df.sort_values("VIF", ascending=False)
print(vif_df)
```

### 변수 그룹핑

```python
# 그룹 기준 정의
count_cols = [col for col in ca_columns if '이용건수' in col or '신청건수' in col]
amount_cols = [col for col in ca_columns if '이용금액' in col or '최대이용금액' in col]
month_cols = [col for col in ca_columns if '이용개월수' in col]
recency_cols = [col for col in ca_columns if '이용후경과월' in col]
last_use_cols = [col for col in ca_columns if '최종이용일자' in col]

# 그룹화된 피처 만들기
df['총_이용건수_CA'] = df[count_cols].sum(axis=1)
df['총_이용금액_CA'] = df[amount_cols].sum(axis=1)
df['총_이용개월수_CA'] = df[month_cols].sum(axis=1)
df['평균_이용후경과월_CA'] = df[recency_cols].mean(axis=1)
df['최종이용일자_CA'] = df[last_use_cols].max(axis=1)
```

```python
# 0인 사람은 '미사용', 나머지는 분위수로 나누기
nonzero = df[df['총_이용금액_CA'] > 0].copy()
nonzero['금액등급'] = pd.qcut(nonzero['총_이용금액_CA'], q=4, labels=['낮음', '중간', '높음', '상위고액'])

# 병합
df['금액등급'] = '미사용'
df.loc[nonzero.index, '금액등급'] = nonzero['금액등급']
```

### 최종 상관계수 및 VIF

```python
# 1️⃣ 총 이용금액 (R3M + R6M + R12M)
df['총_이용금액_CA'] = df[['이용금액_CA_R3M', '이용금액_CA_R6M', '이용금액_CA_R12M']].sum(axis=1)

# 2️⃣ 총 이용건수 + 총 이용개월수
df['총_이용건수_CA'] = df[['이용건수_CA_R3M', '이용건수_CA_R6M', '이용건수_CA_R12M']].sum(axis=1)
df['총_이용개월수_CA'] = df[['이용개월수_CA_R3M', '이용개월수_CA_R6M', '이용개월수_CA_R12M']].sum(axis=1)

# 3️⃣ 평균 이용후 경과월
df['평균_이용후경과월_CA'] = df[['이용후경과월_CA']].mean(axis=1)

# 4️⃣ 월평균 이용건수 (총 이용건수 / 총 이용개월수)
df['월평균_이용건수_CA'] = df['총_이용건수_CA'] / (df['총_이용개월수_CA'] + 1e-5)

# 5️⃣ 월평균 이용금액 (총 이용금액 / 총 이용개월수)
df['월평균_이용금액_CA'] = df['총_이용금액_CA'] / (df['총_이용개월수_CA'] + 1e-5)

# 6️⃣ ATM 신청건수 통합 (B0 + R6M)
df['총_ATM신청건수_CA'] = df[['신청건수_ATM_CA_B0', '신청건수_ATM_CA_R6M']].sum(axis=1)

# 7️⃣ 불필요한 중간 변수 제거
df.drop(columns=['총_이용건수_CA', '총_이용개월수_CA'], inplace=True)  # 월평균 변수로 대체했기 때문
```

```python
from statsmodels.tools.tools import add_constant
from statsmodels.stats.outliers_influence import variance_inflation_factor

# 통합된 변수 리스트
final_vars = ['월평균_이용건수_CA', '월평균_이용금액_CA', '총_이용금액_CA', '평균_이용후경과월_CA', '총_ATM신청건수_CA']

# ⭕ VIF 계산
X = df[final_vars]
X_const = add_constant(X)

vif_df = pd.DataFrame()
vif_df['변수'] = X.columns
vif_df['VIF'] = [variance_inflation_factor(X_const.values, i+1) for i in range(X.shape[1])]
print(" VIF 결과:")
print(vif_df.sort_values('VIF', ascending=False))

# ⭕ 상관계수 행렬 출력
corr_matrix = df[final_vars].corr().round(2)
print("\n상관계수:")
print(corr_matrix)
```

```python
# 분석할 변수 리스트 (예시: 통합 변수들)
cols_to_plot = ['월평균_이용건수_CA', '월평균_이용금액_CA', '총_이용금액_CA', '평균_이용후경과월_CA', '총_ATM신청건수_CA']

# 상관계수 행렬 계산
corr = df[cols_to_plot].corr().round(2)

# 히트맵 시각화
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1, linewidths=0.5)
plt.title(" 통합 변수 간 상관계수 히트맵", fontsize=14)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# 1. 클러스터별 요약할 변수 정의
cluster_vars = [
    '월평균_이용건수_CA',
    '월평균_이용금액_CA',
    '총_이용금액_CA',
    '총_ATM신청건수_CA'
]

# 2. 클러스터 ID → 고객군 이름 매핑 딕셔너리
cluster_map = {
    0: '휴면 고객군',
    1: 'ATM 중심 충성 고객군',
    2: '고빈도·고액 거래군'
}

# 3. 클러스터별 평균값 요약 계산
cluster_summary = (
    df.groupby('고객클러스터')[cluster_vars]
    .mean()
    .round(2)
    .rename_axis('클러스터')
    .reset_index()
)

# 4. 고객군 이름 열 추가
cluster_summary['고객군'] = cluster_summary['클러스터'].map(cluster_map)

# 5. 컬럼 순서 재정렬
cols = ['고객군'] + cluster_vars
cluster_summary = cluster_summary[cols]

# 6. 결과 출력
print("클러스터별 평균 특성 요약 (고객군 이름 포함):")
print(cluster_summary)
```

### 예측 모델 기반 실험

```python
### 교차검증을 통한 모델 안정성
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

#  평균_이용후경과월_CA 제외한 입력 변수
features = ['월평균_이용건수_CA', '월평균_이용금액_CA', '총_이용금액_CA', '총_ATM신청건수_CA']
X = df[features]
y = df['이탈여부']  # 이미 생성된 타깃 변수

# 5-Fold 교차검증 (F1 score 기준)
rf = RandomForestClassifier(random_state=42)
scores = cross_val_score(rf, X, y, cv=5, scoring='f1')

# 결과 출력
print(" F1 Score (5-fold CV):", scores)
print(" 평균 F1:", scores.mean().round(4))

```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.inspection import permutation_importance

# 1. 모델 학습
rf = RandomForestClassifier(n_estimators=50, random_state=42)
rf.fit(X, y) 

# 2. 데이터 일부 샘플링 (속도 최적화)
X_small = X.sample(1000, random_state=42)
y_small = y.loc[X_small.index]

# 3. Permutation Importance 계산 (반복 3회로 속도 개선)
result = permutation_importance(rf, X_small, y_small, n_repeats=3, random_state=42)

# 4. 결과 정리
perm_df = pd.DataFrame({
    '변수': X_small.columns,
    '중요도': result.importances_mean
}).sort_values(by='중요도', ascending=False)

# 5. 시각화
plt.figure(figsize=(7, 4))
plt.barh(perm_df['변수'], perm_df['중요도'], color='darkorange')
plt.title(" Permutation Importance", fontsize=13)
plt.xlabel("중요도 (교란 후 성능 저하)")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
### 다른 모델과 성능 비교 (로지스틱, XGBoost)
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.metrics import f1_score

# Logistic Regression
lr = LogisticRegression(max_iter=1000, random_state=42)
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)
f1_lr = f1_score(y_test, y_pred_lr)

# XGBoost
xgb = XGBClassifier(random_state=42)
xgb.fit(X_train, y_train)
y_pred_xgb = xgb.predict(X_test)
f1_xgb = f1_score(y_test, y_pred_xgb)

print(f" Logistic F1: {f1_lr:.4f}")
print(f" XGBoost F1: {f1_xgb:.4f}")
```
