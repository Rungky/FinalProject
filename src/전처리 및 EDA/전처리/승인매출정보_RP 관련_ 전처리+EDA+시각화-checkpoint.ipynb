### 변수 그룹핑

```python
rp_types = [
    '통신', '아파트', '렌탈', '가스', '전기', '보험',
    '학습비', '유선방송', '건강', '교통', '제휴사서비스직접판매'
]

for t in rp_types:
    try:
        df[f'RP_{t}_총건수'] = df[f'RP건수_{t}_B0M']
        df[f'RP_{t}_최근성'] = 12 - df[f'RP후경과월_{t}']
        df[f'RP_{t}_사용점수'] = df[f'RP_{t}_총건수'] + df[f'RP_{t}_최근성']
        df[f'RP_{t}_사용여부'] = (df[f'RP_{t}_총건수'] > 0).astype(int)
    except KeyError:
        print(f"⚠️ {t} 관련 컬럼 누락 - 건너뜀")
```

```python
summary = pd.DataFrame()

for t in rp_types:
    try:
        summary.loc[t, '사용률'] = df[f'RP_{t}_사용여부'].mean()
        summary.loc[t, '평균점수'] = df[f'RP_{t}_사용점수'].mean()
    except KeyError:
        continue
```

```python
score_cols = [f'RP_{t}_사용점수' for t in rp_types if f'RP_{t}_사용점수' in df.columns]
bin_cols = [f'RP_{t}_사용여부' for t in rp_types if f'RP_{t}_사용여부' in df.columns]

df['RP_총사용점수'] = df[score_cols].sum(axis=1)
df['RP_사용항목수'] = df[bin_cols].sum(axis=1)
```

```python
df['RP_이용여부'] = (df['RP금액_B0M'] > 0).astype(int)
df['RP_이용여부_라벨'] = df['RP_이용여부'].map({0: '미사용', 1: '사용'})
```

### 사용률 1% 미만 컬럼 제거

```python
# 사용여부 컬럼 중 사용률이 1% 미만인 항목 추출
low_usage_bins = [col for col in bin_cols if df[col].mean() < 0.01]

# 사용점수 컬럼으로 매핑
low_usage_scores = [col.replace('사용여부', '사용점수') for col in low_usage_bins]

# 제거 대상 통합
remove_features = low_usage_scores + ['RP_유선방송_사용점수', 'RP_건강_사용점수']

# 최종 사용점수 컬럼에서 제거
filtered_score_cols = [col for col in score_cols if col not in remove_features]
```

### VIF 계산

```python
from statsmodels.tools.tools import add_constant
from statsmodels.stats.outliers_influence import variance_inflation_factor

# 데이터셋 구성
X = df[filtered_score_cols].dropna()
X_const = add_constant(X, has_constant='add')

# VIF 계산
vif = pd.Series(
    [variance_inflation_factor(X_const.values, i) for i in range(X_const.shape[1])],
    index=X_const.columns
)

# 상수항 'const' 제거 (존재할 경우)
if 'const' in vif.index:
    vif = vif.drop('const')

# 정렬 후 출력
vif = vif.sort_values(ascending=False)
display(vif)
```

### 클러스터링

```python
cluster_vars = filtered_score_cols + ['RP_총금액']
cluster_input = df[cluster_vars].dropna()
```

```python
### 최적 클러스터 개수 탐색 (Elbow Method)
from sklearn.cluster import KMeans

inertia = []
K = range(1, 11)

for k in K:
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(cluster_input)
    inertia.append(km.inertia_)

plt.figure(figsize=(8,5))
plt.plot(K, inertia, marker='o')
plt.xlabel('k (클러스터 개수)')
plt.ylabel('Inertia (군집 응집도)')
plt.title('Elbow Method로 최적 k 찾기')
plt.grid(True)
plt.show()

```

```python
### 클러스터링 적용

k = 3  
kmeans = KMeans(n_clusters=k, random_state=42)
df['클러스터'] = kmeans.fit_predict(cluster_input)
```

```python
# 클러스터 라벨 기준 평균표 (가독성을 위한 정렬 포함)
summary = df.groupby('클러스터_라벨')[cluster_vars].mean().round(1)
summary = summary.loc[['전방위 고사용자', '교통 중심 실속파', '저활성 고객군']]
display(summary)
```

### Racar chart로 군집별 특징 시각화

```python
# 사용점수 기준 피처만 추출
features = [col for col in cluster_vars if '총금액' not in col]

# 데이터 재정렬
plot_data = summary[features]

# 각 항목별 최대치로 정규화 (0~1 비율)
plot_data_norm = plot_data / plot_data.max()

# 레이더 차트용 세팅
labels = features
num_vars = len(labels)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]  # 첫 점 마감

fig, ax = plt.subplots(figsize=(6,6), subplot_kw=dict(polar=True))

# 색상 팔레트
colors = ['#ff7f0e', '#1f77b4', '#2ca02c']

# 각 군집 그리기
for idx, (label, row) in enumerate(plot_data_norm.iterrows()):
    values = row.tolist()
    values += values[:1]  # 시작점 반복
    ax.plot(angles, values, label=label, color=colors[idx])
    ax.fill(angles, values, alpha=0.2, color=colors[idx])

# 눈금, 라벨 설정
ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)
ax.set_thetagrids(np.degrees(angles[:-1]), labels, fontsize=10)
ax.set_title('군집별 RP 사용 패턴 비교 (정규화)', fontsize=14, pad=20)
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))
plt.tight_layout()
plt.show()
```
